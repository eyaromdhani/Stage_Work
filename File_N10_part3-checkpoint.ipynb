```python
import pandas as pd
import matplotlib.pyplot as plt


NEW_df['Occurred On (NT)'] = pd.to_datetime(NEW_df['Occurred On (NT)'], errors='coerce')
NEW_df['Occurred On (NT)'] = NEW_df['Occurred On (NT)'].dt.date


NEW_df['Prefix'] = NEW_df['MO Name'].str.extract(r'(^[A-Z]+)')


grouped_df = NEW_df.groupby(['Prefix', 'Occurred On (NT)']).size().unstack(fill_value=0)


color_map = {
    'SBO': 'lightseagreen',
    'ARI': 'gold',
    'TOZ': 'indigo',
    'KEB': 'crimson',
    'SIL': 'darkviolet',
    'MAN': 'seashell',
    'BIZ': 'pink',
    'MON': 'orange',
    'JEN': 'brown',
    'SOU': 'black',
    'MAH': 'lavender',
    'GAF': 'lightsalmon',
    'BAR': 'purple',
    'NAB': 'lightpink',
    'ZAG': 'yellow',
    'TAT': 'red',
    'SFA': 'magenta',
    'KAI': 'cyan',
    'KAS': 'lightgreen',
    'TUN': 'green',
    "any" : "#F0F0F0"
}

tag_regions = {'TUN': 'TUNIS', 'BAR': 'BEN AROUS', 'ARI': 'ARIANA', 'NAB': 'NABEUL', 'MAN': 'MANOUBA', 'BIZ': 'BIZERTE', 'ZAG': 'ZAGHOUAN', 'JEN': 'JENDOUBA', 'TAT': 'TATAOUINE', 'KEB': 'KEBILI', 'TOZ': 'TOUZEUR', 'GAF': 'GAFSA', 'SOU': 'SOUSSE', 'MON': 'MONASTIR', 'MAH': 'MAHDIA', 'SFA': 'SFAX', 'KAI': 'KAIROUAN', 'KAS': 'KASSERINE', 'SBO': 'SIDI BOUZID', 'SIL': 'SILIANA'}


default_color = 'gray'

# Plot for each prefix
prefixes = grouped_df.index
for prefix in prefixes:
    region_counts = grouped_df.loc[prefix]
    
    plt.figure(figsize=(14, 7))
    
    color = color_map.get(prefix, default_color)
    plt.bar(region_counts.index, region_counts.values, color=color, alpha=0.7, label=f'Occurred in {prefix}')
    
    plt.xlabel('Date')
    plt.ylabel('Number of Events')
    plt.title(f'Occurrences per Date in {prefix}', fontdict={'fontsize': 16, 'fontweight': 'bold', 'color': color})
    
    plt.xticks(rotation=45)
    plt.legend()
    plt.tight_layout()
    plt.savefig(f'./plots/{prefix}_plot.png')

    plt.show()

```

```python
import folium
import json
from folium import IFrame
import base64
import os

regions_translate = {'TUNIS': 'TUN',
                     'BEN AROUS': 'BAR',
                     'ARIANA': 'ARI', 
                     'NABEUL': 'NAB', 
                     'MANOUBA': 'MAN', 
                     'BIZERTE': 'BIZ', 
                     'ZAGHOUAN': 'ZAG',
                     'JENDOUBA': 'JEN',
                     'TATAOUINE': 'TAT', 
                     'KEBILI': 'KEB',
                     'TOUZEUR': 'TOZ',
                     'GAFSA': 'GAF',
                     'SOUSSE': 'SOU',
                     'MONASTIR': 'MON',
                     'MAHDIA': 'MAH', 
                     'SFAX': 'SFA', 
                     'KAIROUAN': 'KAI',
                     'KASSERINE': 'KAS',
                     'SIDI BOUZID': 'SBO',
                     'SILIANA': 'SIL',
                    "ZARZIS": "any",
                    "GABES" : "any",
                    "LE KEF": "any"}



# Load GeoJSON data
with open('./maps_final/TN-gouvernorats.geojson', encoding='utf-8') as f:
    tunisia_geojson = json.load(f)

# Initialize the map
m = folium.Map(
    location=[34.0, 9.0],
    zoom_start=5,
    tile="https://{s}.basemaps.cartocdn.com/rastertiles/voyager/{z}/{x}/{y}.png"
)

# Function to create image popups
def add_image_popup(feature):
    region_name = feature['properties']['gouv_fr']
    image_path = f'plots/{regions_translate[region_name.upper()]}_plot.png'
    print(image_path)
    if os.path.exists(image_path):
        with open(image_path, 'rb') as img_file:
            img_data = img_file.read()
        encoded_image = base64.b64encode(img_data).decode()
        image_html = f'<img src="data:image/png;base64,{encoded_image}" width="400" height="300">'
        popup = folium.Popup(IFrame(image_html, width=420, height=320), max_width=420)
        return popup
    else:
        return folium.Popup("No Data Available")


# Add GeoJSON data with popups
for feature in tunisia_geojson['features']:
    color = color_map[regions_translate[feature['properties']['gouv_fr'].upper()]]
    folium.GeoJson(
        feature,
        name='Tunisia Governorates',
        style_function=lambda x, col=color :{
            'fillColor': col,
            'color': 'grey',
            'weight': 1,
            'fillOpacity': 0.5
        },
        popup=add_image_popup(feature)
    ).add_to(m)

folium.LayerControl().add_to(m)

# Save and check the map output
m.save('tunisia_map_with_plots.html')
```
